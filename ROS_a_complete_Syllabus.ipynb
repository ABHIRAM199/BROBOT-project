# ROS [Robot operating System]

# Introduction to ROS

ROS (Robot Operating System) is a flexible framework for writing robot software. It provides tools, libraries, and conventions for developing complex robotic applications. ROS isn't an operating system in the traditional sense but a middleware that allows different parts of a robotic system to communicate and work together.

**Key Features of ROS**

*   **Distributed Computing:** ROS allows various components (called nodes)  
    of a robotic system to run on multiple devices or computers.
*   **Code Reusability:** It provides a wide array of pre-written packages  
    for common robotic functionalities.
*   **Communication Interface:** Nodes communicate via topics, services, or
    actions, enabling modular and scalable design.
*   **Support for Simulation:** ROS integrates with simulation tools like
    Gazebo for testing without real hardware.

#  Components of ROS

ROS comprises several components that together enable robot programming

**Node** : The basic building block of ROS, responsible for specific tasks
           like sensing, computation, or control.
    
**Topic** : Asynchronous communication channel for data exchange between
            nodes.

**Service** : Synchronous request/response communication between nodes.

**Message** : Data structure used in topics, services, and actions.

**Actionlib** : Framework for long-running tasks requiring feedback and
                results.

**ROS Master** : Central registry that enables nodes to find each other and
                 communicate.
          
**Parameter Server** : Global storage for configuration data, accessible by
                       all nodes.





**Nodes**



*   A node is a single process in ROS
*   Nodes are responsible for performing specific tasks, such as reading a
    sensor, controlling a motor, or processing data.

* Example :
1.   A node named `camera_driver` might handle input from a camera.

2.   Another node, `image_processor`, could process that camera feed.

Nodes communicate with each other to achieve complex tasks




            



**Topics**



*   Topics are the communication channels in ROS that enable nodes to  
    exchange messages.

*   Nodes publish data to a topic or subscribe to a topic to receive data.

For example,

A `camera_node` might publish images to a topic named `/camera/image_raw`.

A `vision_node` subscribes to `/camera/image_raw` to process those images.


**Messages**



*   Messages are the data structures used to exchange information between
    nodes via topics
*   A message might contain data like integers, floating-point numbers,
    strings, or even complex structures like images or point clouds.
*   ROS provides many pre-defined message types, such as:




1.   `std_msgs/String`: For text data.
2.   `sensor_msgs/Image`: For camera images.
3.   `geometry_msgs/Twist`: For velocities in linear and angular directions.







**Services**

Services provide a request/response model for communication.
Unlike topics, which are unidirectional, services are synchronous and involve  a client and a server:
* A client sends a request to a service.
* The server processes the request and sends a response back.
A service named `/spawn` in a simulator like Gazebo might take a request to spawn a new robot and return confirmation.

**Actionlib (Actions)**


Actions are used for long-running tasks that might require feedback  
during execution, like moving a robotic arm.

Actions use three types of messages:


1.   **Goal :** The desired end state.
2.   **Feedback :** Progress updates during execution.
3.   **Result :** The final outcome.









**ROS Master**

* The Master is the backbone of the ROS network
* It runs as a single process and provides name registration and lookup
  services to other nodes.

When a node starts, it registers itself with the Master, providing:


1.   Its name
2.   The topics it publishes or subscribes to.
3.   Any services it offers

Once registered, nodes can discover and communicate directly without involving the Master.








**Parameter Server**

The Parameter Server is a shared, multi-node dictionary that nodes use to store and retrieve data. Parameters are typically used for configuration and initialization





*   A global shared dictionary for storing configuration parameters,
    accessible to all nodes
*   Parameters are typically used for static or semi-static settings like
    robot dimensions, sensor calibration values, or operational limits.
*   Nodes can read from and write to the Parameter Server.


Example:


1.   `rosparam set /robot_name "MyRobot"`: Set a parameter.
2.   `rosparam get /robot_name`: Get the value of a parameter
3.   `rosparam list`: List all parameters
